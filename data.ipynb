The data collected from Foursquare is very useful in this regrads. The data lists famous spots around a pre-defined area to the Airport, with dedicated properties to each spot containing its location (latitude and longitude), a unique id, name and category, image of the spot, popular item and many more. This data can utilized in so many ways to increase interest to check out these spots.

The properties used in this project are name, categories, latitude and longitude. Other properties can be added to make it more flexible or useful according to the need.
