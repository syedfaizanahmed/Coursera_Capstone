# METHODOLOGY

In this section, I will detail the methodology used to collect the data and map onto the cities.

The main part of this project is to collect the data for the venues in the aforementioned cities airports. The data is collected through Foursquare, which is one of the best databases for developers, used across the world for its simplicity and details in data collection and representation. The data stored in the Foursquare databases include names and categories of venues in cities, their locations, images, main items that the venue is famous for, customer ratings and comments. Database also include customers data, which could be used to identify the customer interests and likenesses. Each spot and customer is given a unique id that could be used by the developers to identify each customer and venue. 

For this project, the data collected had name, categories and its locations for the venues in close proximity to the international airports in the three cities of Karachi, Lahore and Islamabad. The data was collected throug using the unique developer client id and password id, which could be used to call 1500 calls per day through the Foursquare database. 

At first the geo-location of the cities was collected using geolocator function in the geopy package in python. This package has some exciting features including geolocation of cities and towns across the globe. The Function geolocator was used recall the latitudes and longitudes of Karachi, Lahore and Islamabad, and centered around the citys' International airports, namely Jinnah International, Allama Iqbal International and Islamabad International Airport. 

This location for the citys' airports was then used to find the venues and spots close to the airports by calling from Foursquare database and storing in pandas dataframe. Each of these dataframe was cleaned and wrangled to refine it as required. The resulting dataframes included venue name, category, latitude and longitude as shown in one of the tables above.   

The categories of the venues and their count varies for each city, which are also showed in the bar plots and tables above. 

The next step was to create the might highlighting these venues in the cities. Folium package was used for this purpose. Folium package is one of the most widely used package for the rendering of interactive maps and images on python. The location of citys' airports extracted from Foursquare database was used to create map through Folium and identify the airport with a red circle. Then, the databases for venues of each city are used to impose the venues on the same city map with blue dots highlighting the venues while airport as red circle. This is done to give a visual approximation of the distance between the airport and the venue.

The following image shows the map of Karachi with venues around its International Airport.

```python
khi_venues_map = folium.Map(location=[khi_latitude, khi_longitude], zoom_start=13) # generate map centred around the Airport

# add a red circle marker to represent the 
folium.CircleMarker(
    [khi_latitude, khi_longitude],
    radius=10,
    color='red',
    popup='Karachi International Airport',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(khi_venues_map)

# add the Italian restaurants as blue circle markers
for lat, lng, label in zip(khi_nearby_venues.lat, khi_nearby_venues.lng, khi_nearby_venues.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(khi_venues_map)

# display map
khi_venues_map
```

# DISCUSSION

In the maps shown above and the charts earlier, it can be show that there are some commonalities and some differences among the three airports and their surroundings. First of all, among the commonalities, it can be seen that the most close venues for each airport has category related to cuisine. It is not very surprising as food restaurants and dining hotels are the most profitable businesses in these megacities. It is also not surprising that other venues like gym/sports activities, cafe's and clubs, and shopping and cinema halls are also very close to the airports. It shows that the cities are vibrant and full of fun activities that can be very interesting for tourists visiting the cities.

Among differences, it can be seen that the venues around Karachi airport are spread all around it while in Lahore and Islamabad airports, the venues are much on one side of the airports. This is due to the fact that teh Karachi Airport is one of the oldest airports in the sub-continental region which was initially built right outside the city limits but the over the years, city has expanded beyond the airport and it now sits almost in the middle of the city.

Airports in Lahore and Islamabad are relatively new and are right outside the city. which is why most of the venues are on one side of the airport. For Lahore, the main city and its venues are west to the airport while for Islamabad, the venues are to the east direction of the airport.

This provides a unique look at the cities and ease of going to these venues for the tourists. In Karachi, the tourist can go any direction of the airport and will be easy to find food restaurants and hotels. While tourists in Lahore and Islamabad airports, would have to travel a little farther to arrive to their desired location.
